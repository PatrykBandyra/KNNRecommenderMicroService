```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
users = pd.read_json("../data/raw/users.jsonl", lines=True)
sessions = pd.read_json("../data/raw/sessions.jsonl", lines=True)
products = pd.read_json("../data/raw/products.jsonl", lines=True)
```

```python
category_list = products['category_path'].unique()
category_list
```

```python
def one_hot_encode(element, list):
    one_hot_encode_list = []
    
    for e in list:
        if element == e:
            one_hot_encode_list.append(1)
        else:
            one_hot_encode_list.append(0)
    return one_hot_encode_list
```

```python
products['category_path'] = products['category_path'].apply(lambda x: one_hot_encode(x, category_list))
products
```

```python
products['price'] = (products['price'] - products['price'].min()) / (products['price'].max() - products['price'].min())
products['user_rating'] = (products['user_rating'] - products['user_rating'].min()) / (products['user_rating'].max() - products['user_rating'].min())
products
```

```python
for index, product in products.iterrows():
    print(product['product_name'], product['product_id'])
```

```python
from scipy import spatial
def similarity(product_id1, product_id2):
    a = products.iloc[product_id1]
    b = products.iloc[product_id2]
    
    categoryA = a['category_path']
    categoryB = b['category_path']
    category_distance = spatial.distance.cosine(categoryA, categoryB)
    
    priceA = a['price']
    priceB = b['price']
    price_distance = abs(priceA - priceB) * 1
    
    ratingA = a['user_rating']
    ratingB = b['user_rating']
    rating_distance = abs(ratingA - ratingB)
    
    return category_distance + price_distance + rating_distance
```

```python
import operator
def get_distances(product_id):
    p = products.index[products['product_id'] == product_id][0]
    distances = []
    
    for index, product in products.iterrows():
        if product['product_id'] != product_id:
            dist = similarity(index, p)
            distances.append(dist)
        else:
            distances.append(0)
    
    return distances

def get_neighbours(distances, K):
    distances = [(index, dist) for index, dist in enumerate(distances)]
    distances.sort(key=operator.itemgetter(1))
    neighbours = []
    
    for x in range(K):
        neighbours.append(distances[x])
    return neighbours
```

```python
neighbours = get_neighbours(get_distances(1002), 5)
for neighbour in neighbours:
    print(products.iloc[neighbour[0]])
    print(f"Distance: {neighbour[1]}")
```

```python
all_distances = []
for i in range(len(products)):
    all_distances.append(get_distances(products['product_id'].iloc[i]))
```

```python
def get_recomendation(user_id, K, all_distances):
    s = sessions.index[sessions['user_id'] == user_id].tolist()
    ids = []
    for i in s:
        ids.append(products.index[products['product_id'] == sessions['product_id'].iloc[i]][0])
    
    distances = [0] * len(products)
    for id in ids:
        distances = [sum(x) for x in zip(all_distances[id], distances)]
    
    maxi = max(distances)
    for id in ids:
        distances[id] = maxi
    
    return get_neighbours(distances, K)
```

```python
neighbours = get_recomendation(107, 5, all_distances)
for neighbour in neighbours:
    print(products.iloc[neighbour[0]])
    print(f"Distance: {neighbour[1]}")
```
